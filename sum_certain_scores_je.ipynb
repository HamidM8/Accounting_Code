```python
import pandas as pd
df = pd.DataFrame({"ITEM_NUM": ["1", "2", "1","2","3"],\
    "DEBIT_CREDIT": ["50", "CA", "EW","40","54"],\
    "GL_ACCOUNT": ["50000", "56000", "65000","70000", "45000"],\
    "AMOUNT_TR": [45000,45000,5000,4000,10000],
    'DOC_NUM': ["100", "100", "102", "102","102"], 
    'Risk_ERP': [0,0,1,0,1], 
    'Risk_Control': [1,1,1,0,1], 
    'Risk_Error': [0,1,0,1,1],  
    })
df
```

```python
Risk_cols = []
for col in df.columns.tolist():
    if col.split("_")[0]=="Risk":
        Risk_cols.append(col)

print(Risk_cols)
```

```python
df["score_Final"] = df[Risk_cols].sum(axis=1)
df
```
